### Final scraping code

```python
# EXTRA INFO
url = 'https://www.ah.nl/zoeken?query=kip&page=1'
page = requests.get(url)

soup = BeautifulSoup(page.content, 'html')

AH_products = soup.find_all("article", class_="product-card-portrait_root__sZL4I product-grid-lane_gridItem__eqh9g") #get all the HTML fragments with descriptions
AH_data = []

num = 0
for product in AH_products:

    # Reading from the html element on the results page
    Product_name = product.find("a", class_="link_root__65rmW product-card-portrait_link__1ROKP").get_text()
    Product_price = product.find("div", {"class":["price-amount_root__37xv2 price-amount_highlight__3WjBM price_amount__2Gk9i price_highlight__3B97G", "price-amount_root__37xv2 price-amount_highlight__3WjBM price-amount_bonus__27nxZ price_amount__2Gk9i price_highlight__3B97G"]}).get_text()
    Product_quantity = product.find("span", class_="price_unitSize__8gRVX").get_text()
    
    link = product.find("a", class_= "link_root__65rmW").attrs["href"]
    print(link)
    completeLink = f"https://www.ah.nl{link}"
    
    productpage = requests.get(completeLink)
    soupProduct = BeautifulSoup(productpage.content, 'html') # Product HTML page

    # Reading from the product html page. The scraping will read the class, if there's nothing, it will leave it empty.
    
    try:
        Product_summary = soupProduct.find("div", class_= "product-summary_root__3_CJs").get_text()
    except:
        Product_summary = ""
        
    try:
        Product_description = soupProduct.find("ul", class_="product-info-description_list__MUNdA").get_text()

    except:
        Product_description = ""
        
    try:
        Product_ingredients = soupProduct.find("p", class_="typography_root__18FkK typography_variant-paragraph__33rgM typography_hasMargin__26L1z").get_text()

    except:
        Product_ingredients = ""       

    AH_list = {"Product name": Product_name, "Description": Product_description, "Summary":Product_summary, "Price": Product_price, "Ingredients": Product_ingredients,"Quantity": Product_quantity} 
    AH_data.append(AH_list)
    num += 1
    print(AH_data)
    print(num)
    

# Here we turn it into a json file
    
with open("kip_extra.json", "w") as json_file:
    json.dump(AH_data, json_file) 
    
    
# Here we turn it into a dataframe
    
"""
data_items = articles_data.items()
data_list = list(data_items)
"""

kip_df = pd.DataFrame(AH_data)

# Add new column corresponding to the type of food

kip_df ["key word"] = "Chicken"
#print (peulvrucht_df)

# Export to CSV

kip_df.to_csv("kip_extra.csv")
    
```
